### outliers in min_payment_amt, probability_of_full_payment

1. Outlier in the min_payment should be treated since there is a significant difference in the min and max.
   treat this with mean.
2. Outlier in the probablity_of_full_payment is very marginal we can treat this with mean


```python
def outlier_treat (col):
    sorted(col)
    Q1,Q3= col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5*IQR)
    upper_range= Q3+ (1.5*IQR)
    return lower_range, upper_range
```

```python
lrminpay, urminpay=outlier_treat (df['min_payment_amt'])
df['min_payment_amt']=np.where(df['min_payment_amt']>urminpay,urminpay,df['min_payment_amt'])
df['min_payment_amt']=np.where(df['min_payment_amt']<lrminpay,lrminpay,df['min_payment_amt'])

lrfullpay, urfullpay = outlier_treat (df['probability_of_full_payment'])
df['probability_of_full_payment']=np.where(df['probability_of_full_payment']>urfullpay,urfullpay,df['probability_of_full_payment'])
df['probability_of_full_payment']=np.where(df['probability_of_full_payment']<lrfullpay,lrfullpay,df['probability_of_full_payment'])

df.shape
```
