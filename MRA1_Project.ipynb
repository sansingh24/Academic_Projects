```python
numcol= df[['SALES','DAYS_SINCE_LASTORDER','PRICEEACH','QUANTITYORDERED']]
```

```python
df.describe().T
```

```python
def univariateanalysis (column,bins):
    print ('Description of '+ column)
    print('------------------------------------------------------------------')
    print (df[column].describe(), end='')
    
    plt.figure()
    print ('Skewness of' + column)
    print ('-----------------------------------------------------------------')
    sns.distplot(df[column],kde= True, color ='r');
    plt.show()
    
    plt.figure()
    print ('Boxplot of '+ column)
    print ('----------------------------------------------------------------')
    ax= sns.boxplot (x=df[column])
    plt.show()
    
```

```python
for i in numcol:
    univariateanalysis(i,16)
```

```python
catcol=df[['PRODUCTLINE', 'MSRP','PRODUCTCODE','COUNTRY','ORDERNUMBER','CUSTOMERNAME']]
```

```python
def univariatenalysiscat (column,bins):
    plt.figure()
    print('Countplot of' + column)
    print('------------------------------------------------------------------------------')
    ax= sns.countplot(x=df[column])
    plt.show()
    
    print ('Unique Values of '+ column)
    print ('-----------------------------------------------------------------------------')
    print (df[column].value_counts(), end='')
```

```python
for i in catcol:
    univariatenalysiscat(i,16)
```
