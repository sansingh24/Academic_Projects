```python
returns = np.log(data.drop(['Date','Date'],axis=1)).diff(axis = 0, periods = 1) 
```

```python
returns.head()
```

```python
returns['Infosys'].mean()
```

```python
stock_means = returns.mean(axis = 0)
stock_means
```

```python
stock_sd = returns.std(axis = 0)
stock_sd
```

```python
Infer_data = pd.DataFrame({'Average':stock_means, 'Volatility': stock_sd})
Infer_data 
```

```python
plot = sns.scatterplot(Infer_data['Volatility'], Infer_data['Average'])
plot.axvline(x=0.020257,linestyle='--', color ="green")
plot.axhline(y=0.000683,linestyle='--', color = "green")
plt.show()
```
