```python
# checking for the columns with outliers for treatment

plt.figure(figsize=(18,12))
numcol.boxplot(vert=0)
```

```python
def treat_outlier(col):
    sorted(col)
    Q1,Q3 = np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range = Q1-(1.5*IQR)
    upper_range= Q3+(1.5*IQR)
    return lower_range, upper_range
```

```python
for i in numcol:
    lr,ur= treat_outlier(numcol[i])
    numcol[i]=np.where(numcol[i]>ur,ur,numcol[i])
    numcol[i]=np.where(numcol[i]<lr,lr,numcol[i])
```

```python
# checking to see if the outliers are treated
plt.figure(figsize=(18,12))
numcol.boxplot(vert=0)
```
